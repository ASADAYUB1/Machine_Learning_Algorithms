```python
from sklearn.cluster import KMeans
```

```python
df.shape
```

```python
df.iloc[:,1:10]
```

```python
data = df.iloc[:,1:10]
```

```python
WCSS= []

for i in range(1,20):
    km = KMeans(i)
    km.fit(data)
    WCSS.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,20), WCSS,marker='o')
```

```python
km = KMeans(n_clusters=5)
y_means = km.fit_predict(data)
```

```python
y_means
```

```python
data['Cluster']=y_means
data
```

```python
import seaborn as sns
sns.scatterplot(x=data.gdpp,y=data.income,hue=data.Cluster,palette='tab10')
```
