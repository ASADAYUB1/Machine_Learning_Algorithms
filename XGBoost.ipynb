# XGBoost   Extreme Gradient Boost

```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('Boston.csv')
```

```python
df.shape
```

```python
X = df.iloc[:, 0:13]
y = df.iloc[:, -1]
```

```python
print('X shape is' , X.shape)
print('y shape is' , y.shape)
```

```python
from xgboost import XGBRFRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
xgb = XGBRFRegressor(eta = 0.9, gamma = 20)
gb = GradientBoostingRegressor()
rf = RandomForestRegressor()
```

```python
from sklearn.model_selection import cross_val_score
```

```python
print('XGB', np.mean(cross_val_score(xgb, X, y, cv= 10)))
print('gb', np.mean(cross_val_score(gb, X, y, cv= 10)))
print('RF', np.mean(cross_val_score(rf, X, y, cv= 10)))
```

## Applying XGBoost on Titanic Dataset

```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv('Titanic_train.csv')
df
```

```python
df.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'], inplace=True)
df
```

```python
X = df.iloc[:,1:]
y = df.iloc[:,0]
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=23)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
```

```python
trf1 = ColumnTransformer([
    
    ('impute', SimpleImputer(),[2]),
    ('impute2', SimpleImputer(strategy='most_frequent'),[6])
    
], remainder='passthrough')
```

```python
trf1.fit_transform(X_train)
```

```python
trf2 = ColumnTransformer([
    
    ('ohe', OneHotEncoder(sparse=False, handle_unknown='ignore'),[1,6])
    ])
```

```python
from xgboost import XGBClassifier
```

```python
clf = XGBClassifier()
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline([
    ('step1',trf1),
    ('step2', trf2),
    ('step3', clf)
])
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```
